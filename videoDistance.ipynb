```python
import cv2
import numpy as np
from cv2 import imread
from cv2 import imshow
import math as m
from scipy.spatial import distance as dist
```

```python
def midpoint(ptA, ptB):
	return ((ptA[0] + ptB[0]) * 0.5, (ptA[1] + ptB[1]) * 0.5)
```

```python
fullBody=cv2.CascadeClassifier('haarcascade_fullbody.xml')
```

```python
cap = cv2.VideoCapture(0) #webcam object
cap.set(3,640) #defining the width
cap.set(4,480)#defining the height
cap.set(10,100)#defining brightness
```

```python
while True:
    success, img = cap.read()
    
    imgGray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
    
    D=0
    fullbody = fullBody.detectMultiScale(imgGray,1.1,4)
    f = 0
    xA=0
    yA=0
    w1=0
    h1=0
    for (x, y, w, h)  in fullbody:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    for (x, y, w, h)  in fullbody:
        if f == 1:
            break
    #     (tl, tr, br, bl) = box
        tl=[x,y+h]
        tr=[x+w,y+h]
        br=[x+w,y]
        bl=[x,y]
        xA=x
        yA=y
        w1=w
        h1=h
        (tlblX, tlblY) = midpoint(tl, bl)
        (trbrX, trbrY) = midpoint(tr, br)
        D = dist.euclidean((tlblX, tlblY), (trbrX, trbrY))
        f = f + 1
    
    
    box = cv2.rectangle(img,(xA,yA),(xA+w1,yA+h1),(255,0,0),2)
    cX = (xA+(w1/2))
    cY = (yA+(h1/2))
    xA=cX
    xA1=xA+w1
    yA=cY
    if(w1 != 0):
        refObj = (box, (cX, cY), D / w1)
    
    
    f=0
    xB=0
    yB=0
    for (x, y, w, h)  in fullbody:
    #     if f != 0:
            xB = (x+(w/2))
            yB = (y+ (h/2))
            D = dist.euclidean((xA, yA), (xB,yB)) / refObj[2]
            D1 = dist.euclidean((xA,yA), ((xA+w1), yA)) / refObj[2]
            print(D1)
            if D == 0:
                continue
            cv2.line(img, (int(xA), int(yA)), (int(xB), int(yB)),color=(176,255,124), thickness=2)
            (mX, mY) = midpoint((xA, yA), (xB, yB))
            D=abs(D-D1)
            D = D/39.37
            if(D < 2):
                cv2.putText(img, "Not following Social Distancing".format(D), (0,0),cv2.FONT_HERSHEY_SIMPLEX, 0.55, color=(1,122,215),thickness=2)
            cv2.putText(img, "{:.1f}m".format(D), (int(mX), int(mY - 10)),cv2.FONT_HERSHEY_SIMPLEX, 0.55, color=(1,122,215),thickness=2)
    
    
    cv2.imshow('Video',img)

    
    
    
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
# After the loop release the cap object
cap.release()
# Destroy all the windows
cv2.destroyAllWindows()
```
